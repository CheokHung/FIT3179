```python
import pandas as pd
df = pd.read_csv("data/athlete_events.csv")
df2 = df[["NOC","Medal"]].copy()
df2
```

```python
countofmedals = df2.groupby(["NOC"]).count()
print(countofmedals.idxmax(), countofmedals.max())
```

```python
nocnames = pd.read_csv("data/noc_regions.csv")
nocnames
```

```python
new = countofmedals.merge(nocnames,on="NOC",how="left")[["region","Medal"]]
new
```

```python
new.to_csv("data/cleaned_data.csv",index=False)
```
