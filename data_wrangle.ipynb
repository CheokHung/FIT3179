```python
import pandas as pd
import numpy as np
df = pd.read_csv("data/athlete_events.csv")
df2 = df[["NOC","Medal","Year"]].copy()
df2
```

```python
nocnames = pd.read_csv("data/noc_regions.csv")
nocnames
```

```python
new = df2.merge(nocnames,on="NOC",)[["region","Medal"]]
new.rename(columns={"region":"Country"},inplace=True)

new
```

```python
countofmedals = new.groupby(["Country"]).count().reset_index()

# countofmedals.loc[countofmedals.Country=="USA","Country"] = "United States of America"
countofmedals
```

```python
countofmedals.Medal.max()
# countofmedals
```

```python
bronze = new[new.Medal=="Bronze"]
bronze_count = bronze.groupby(["Country"]).count()
bronze_count.rename(columns={"Medal":"Bronze"},inplace=True)
# bronze_count
```

```python
silver= new[new.Medal=="Silver"]
silver_count = silver.groupby(["Country"]).count()
silver_count.rename(columns={"Medal":"Silver"},inplace=True)
# silver_count
```

```python
gold =  new[new.Medal=="Gold"]
gold_count = gold.groupby(["Country"]).count()
gold_count.rename(columns={"Medal":"Gold"},inplace=True)
# gold_count
```

```python
countofmedals = new.groupby(["Country"]).count().reset_index()
countofmedals = countofmedals.merge(bronze_count,how="outer",on= "Country",)
countofmedals = countofmedals.merge(silver_count,how="outer",on= "Country",)
countofmedals = countofmedals.merge(gold_count,how="outer",on= "Country",)
countofmedals = countofmedals.fillna(0)
countofmedals["Bronze"] = countofmedals["Bronze"].astype(int).astype(str)
countofmedals["Silver"] = countofmedals["Silver"].astype(int).astype(str)
countofmedals["Gold"] = countofmedals["Gold"].astype(int).astype(str)
countofmedals.loc[countofmedals.Country=="USA","Country"] = "United States of America"
countofmedals
```

```python
countofmedals.to_csv("data/cleaned_data.csv",index=False)
```

```python
new2 = df2.merge(nocnames,on="NOC",)[["region","Medal","Year"]]
new2.rename(columns={"region":"Country"},inplace=True)
```

```python
culmulativemedals =new2.groupby(["Country","Year"],as_index=False).count().assign(
        # assign the cumulative sum of each name as a new column
        cumulative_sum=lambda x: x.groupby('Country')['Medal'].cumsum()
    )
# culmulativemedals
culmulativemedals.loc[culmulativemedals.Country=="USA","Country"] = "United States of America"
culmulativemedals.to_csv("data/yearly_cleaned_data.csv",index=False)
```
