```python
new2 = df2.merge(nocnames,on="NOC",)[["region","Medal","Year"]]
new2.rename(columns={"region":"Country"},inplace=True)
```

```python
culmulativemedals =new2.groupby(["Country","Year"],as_index=False).count().assign(
        # assign the cumulative sum of each name as a new column
        cumulative_sum=lambda x: x.groupby('Country')['Medal'].cumsum()
    )
# culmulativemedals
culmulativemedals.loc[culmulativemedals.Country=="USA","Country"] = "United States of America"
culmulativemedals.to_csv("data/yearly_cleaned_data.csv",index=False)
```

```python
df3 = df[["Medal","Sex"]].copy()
# df3 =  df3.merge(nocnames,on="NOC",)[["region","Medal"]]
# df3.rename(columns={"region":"Country"},inplace=True)
final = df3.groupby(["Sex"],as_index=False).count()
BronzeG = df3[df3["Medal"]=="Bronze"].groupby(["Sex"],as_index=False).count()
BronzeG.rename(columns={"Medal":"Bronze"},inplace=True)
final = final.merge(BronzeG,how="outer",on="Sex")

SilverG = df3[df3["Medal"]=="Silver"].groupby(["Sex"],as_index=False).count()
SilverG.rename(columns={"Medal":"Silver"},inplace=True)
final = final.merge(SilverG,how="outer",on="Sex")

GoldG = df3[df3["Medal"]=="Gold"].groupby(["Sex"],as_index=False).count()
GoldG.rename(columns={"Medal":"Gold"},inplace=True)
byGenderfinal = final.merge(GoldG,how="outer",on="Sex")

byGenderfinal.to_csv("data/cleaned_GroupByGender_data.csv",index=False)
byGenderfinal
```

```python

df4 = df[["Age","Medal"]].copy()
final = df4.groupby(["Age"],as_index=False).count()

BronzeA = df4[df4["Medal"]=="Bronze"].groupby(["Age"],as_index=False).count()
BronzeA.rename(columns={"Medal":"Bronze"},inplace=True)
final = final.merge(BronzeA,how="outer",on="Age")

SilverA = df4[df4["Medal"]=="Silver"].groupby(["Age"],as_index=False).count()
SilverA.rename(columns={"Medal":"Silver"},inplace=True)
final = final.merge(SilverA,how="outer",on="Age")

GoldA = df4[df4["Medal"]=="Gold"].groupby(["Age"],as_index=False).count()
GoldA.rename(columns={"Medal":"Gold"},inplace=True)
byAgefinal = final.merge(GoldA,how="outer",on="Age").fillna(0)
byAgefinal.to_csv("data/cleaned_age_data.csv")
byAgefinal
```
